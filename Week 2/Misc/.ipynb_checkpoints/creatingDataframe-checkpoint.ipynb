```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib notebook
```

```python
#creating a dataframe
#df = pd.DataFrame(np.random.normal(loc=0, scale=1, size=12))
df = pd.DataFrame(np.random.randn(10,4), columns=["a","b","c","d"])
df
```

```python
df.plot.scatter(x="a", y="b", color = "g", s=150, title="Interaction between a and b\n", label ="a and b")
# use \n to add space between the title and the graph
```

```python
#ploting from the dataframe
df.plot.bar()
#to stack the bars and make them horizontal, use df.plot.barh(stacked=True)
plt.show()
```
