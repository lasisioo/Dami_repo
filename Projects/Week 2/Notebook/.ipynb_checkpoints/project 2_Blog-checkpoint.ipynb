### Let's start with some rudimentary cleaning!

The data frame looks a little messy. We need to do some cleaning! To clean the data, we start off by re-naming the heading for the weeks columns:

```python
def cleanVal(val):
    if "week" in val.lower():
        valsplitted = val.split(".")
        newString = valsplitted[0] 
        newList = newString.replace("x","_")
        newNewList = newList.replace("st","").replace("nd","").replace("rd","").replace("th","")
        newVal = valsplitted[1] + " " + newNewList
        return newVal
    return val
finalVal = [cleanVal(x) for x in db.columns.tolist()]
db.columns = finalVal
db.head()
```

##### Using Pandas' built in `melt` function, pivot the weekly ranking data to be long rather than wide. As a result, you will have removed the 72 'week' columns and replace it with two: Week and Ranking. There will now be multiple entries for each song, one for each week on the Billboard rankings.

Now, I want to change the shape of the data frame. I want to remove the 72 'week' columns and replace them with two (week and ranking). This way, the data frame is long rather than wide. Each  Also, I'll drop rows with no data entries 

```python
def cleanVal(val):
    if "week" in val.lower():
        valsplitted = val.split("_") 
        newL = int(valsplitted[1]) 
        return newL
    return val
finalVal2 = [cleanVal(x) for x in db.columns.tolist()]
db.columns = finalVal2

colA = db.columns[0:7].tolist()
colB = db.columns[7:].tolist()

db_1 = pd.melt(db, id_vars=(colA), var_name= "week", value_vars=(colB), value_name= "ranking")
db1 = db_1.sort_values(["track","week"], ascending=True)
db2 = db1.dropna(axis=0, how='any')
db2
```

```python
db2.dtypes
```

The table below shows the description and the types of the column data:

Variable|Description|Type of Variable

```python
newData = db.iloc[:,:7]
allweeks = db.iloc[:,7:]
highestRank = []
for i in range(len(allweeks)):
    highestRank.append((min(allweeks.iloc[i,:])))
newData["HighestRank"] = highestRank

newData["entryDate"] = pd.to_datetime(newData["date.entered"])
newData["peakDate"] = pd.to_datetime(newData["date.peaked"])
#newData["weeksToPeak"] = ((newData["peakDate"] - newData["entryDate"])/7).dt.days #MG Yours
newData["weeksToPeak"] = ((newData["peakDate"] - newData["entryDate"]).dt.days/7).astype(int) #MG Mine
#MG Even though your method gets the same answer, it's a little confusing the was you're getting it.
del newData["entryDate"], newData["peakDate"]
newData
```
